```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 準備模擬的資料

做一條直線: $$ f(x) = 1.2x + 0.8 $$

```python
# 準備50個點
x = np.linspace(0, 5, 50)
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
```

```python
plt.scatter(x, y)
plt.plot(x, 1.2*x+0.8, 'r')
```

# 用sklearn 做線性回歸

這裡要注意我們本來的x 是 $$ [x_{1}, x_{2}, ..., x_{50}] $$

但現在sklearn 的input 要的是 $$ [[x_{1}], [x_{2}], ..., [x_{50}]] $$

這樣明確分割的。

```python
from sklearn.linear_model import LinearRegression
```

```python
# 開一台空的線性回歸機
regr = LinearRegression()
```

```python
X = x.reshape(50, 1)
```

改變array 形狀的方法: array 的變數名.reshape(新的形狀)

```python
# 讓線性回歸機學習
regr.fit(X, y)
```

```python
Y = regr.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, Y, 'r')
plt.plot(x, 1.2*x+0.8, 'g')
```
