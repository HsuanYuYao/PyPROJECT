```python
df = web.DataReader("AAPL", "yahoo", start="2012-9-1", end="2017-8-31")
```

## 計算報酬率:  $\frac{P_{t}-P_{t-1}}{P_{t-1}}$

```python
P.diff()
```

```python
r = P.diff()/P
```

```python
r.plot()
```

```python
r[-100:].plot()
```
