# Classify the quality of red wine based on physicochemical variables
## Discription
Inputs include totally 11 variables: fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates and alcohol. Our goal is to predict whether the sample has a high quality(i.e. quality score>5) or not(i.e. quality score<=5). We first divide the samples into training datas and testing datas. Each time we choose any two of the input variables to perform SVC on training datas, predict the classes which the testing datas belong to and calculate the accuracy score of our model.
## Implementation

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("winequality-red.csv", sep = ";")
```

[dataset](https://archive.ics.uci.edu/ml/datasets/Wine+Quality)

```python
df.head()
```

```python
# check for missing datas
df.isnull().any()
```

```python
# check the data types
df.dtypes
```

```python
# choose two of the 11 input variables
x = df.loc[:, ['fixed acidity','alcohol']].copy().values
x = x.reshape(1599, 2)
```

```python
# two classes
y = df.loc[:, ['quality']].copy().values
y = y.ravel()
z = np.zeros(y.size)
z = z.ravel()
z[y>5] = 1
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
# divide the samples into training datas and testing datas
x_train, x_test, z_train, z_test = train_test_split(x, z, test_size = 0.3, random_state = 1)
```

```python
# plot training datas
plt.scatter(x_train[:, 0], x_train[:, 1], c = z_train)
```

```python
# training SVC model
clf = SVC()
clf.fit(x_train, z_train)
```

```python
# predict the classes which the testing datas belong to
z_predict = clf.predict(x_test)
```

```python
# plot the predicting result
plt.scatter(x_test[:, 0], x_test[:, 1], c = z_predict)
```

```python
# plot the true value
plt.scatter(x_test[:, 0], x_test[:, 1], c = z_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# calculate the accuracy score
accuracy_score(z_test, z_predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(z_test, z_predict)
```

After running all possible choices of the two variables, we found that the combination of fixed acidity and alcohol gives the highest accuracy score, so we show the result of the model with this two input variables above.
