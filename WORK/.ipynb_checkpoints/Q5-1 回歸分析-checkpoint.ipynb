```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib as mpl
```

```python
mpl.rc('font', family='Noto Sans CJK TC')
```

```python
df = pd.read_csv('student_height.csv')
```

[資料下載網址](https://data.gov.tw/dataset/6283)

```python
df.head()
```

```python
x = df.學年度.values
y = df.年齡.values
X = df.loc[:, ['學年度', '年齡']].copy().values
z = df.女.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(X, z)
```

```python
Z = regr.predict(X)
```

```python
x1 = np.reshape(x, (11, 10))
y1 = np.reshape(y, (11, 10))
Z1 = np.reshape(Z, (11, 10))
```

```python
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)
ax.plot_surface(x1, y1, Z1, color='palegreen', alpha=0.5)

ax.set_xlabel('學年度')
ax.set_ylabel('年齡')
ax.set_zlabel('女生平均身高')
```
