```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider, IntSlider
```

```python
#Keras functions
from keras.models import Sequential, Model
from keras.layers import Input, Lambda, concatenate
from keras.layers import Dense
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Conv2DTranspose, UpSampling2D
from keras.optimizers import SGD, Adam
from keras import metrics

#Keras dataset
from keras.datasets import mnist

#Keras utilis function
from keras.utils import np_utils

from keras import backend as K
```

```python
(x_train0, y_train0), (x_test0, y_test0) = mnist.load_data()

print('There are %d training data with size %d x %d' %x_train0.shape)
print('There are %d testing  data with size %d x %d' %x_test0.shape)
```

```python
x_train = x_train0.reshape(60000, -1)
x_test = x_test0.reshape(10000, -1)

x_train -= x_train.min()
x_train = x_train/x_train.max()

x_test -= x_test.min()
x_test = x_test/x_test.max()
```

```python
enc_1 = Dense(100, activation='sigmoid')
enc_mean = Dense(2)
enc_log_var = Dense(2)

dec_2 = Dense(100, activation='sigmoid')
dec_1 = Dense(784, activation='sigmoid')

x = Input(shape=(784,))
enc_x = enc_1(x)

z_mean = enc_mean(enc_x)
z_log_var = enc_log_var(enc_x)
```

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(2,), mean=0, stddev=1)
    return z_mean + K.exp(z_log_var / 2) * epsilon
```

```python
z = Lambda(sampling, output_shape=(2,))([z_mean, z_log_var])
```

```python
dec_x = dec_2(z)
x_hat = dec_1(dec_x)
```

```python
VAE = Model(x, x_hat)
VAE.summary()
```

```python
def vae_loss(x, x_recon):
    recovery_loss = 784 * metrics.binary_crossentropy(x, x_recon)
    kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis = -1)
    return recovery_loss + kl_loss
```

```python
VAE.compile(loss=vae_loss, optimizer=Adam())
```

```python
VAE.load_weights('VAE_handwriting_model_weights.h5')
```

```python
VAE_Encoder = Model(x, z_mean)

VAE_Encoder.summary()
```

```python
h_input = Input(shape=(2,))

VAE_Decoder = Model(h_input, dec_1(dec_2(h_input)))

VAE_Decoder.summary()
```

```python
indices = np.random.randint(x_test.shape[0], size=1000)
VAE_latents = VAE_Encoder.predict(x_test[indices])
```

```python
plt.scatter(VAE_latents[:, 0], VAE_latents[:, 1], c=y_test0[indices], cmap='tab20')
plt.colorbar()
plt.show()
```

```python
def normalized(x):
    x -= x.min()
    x /= x.max()
    return x
```

```python
n = 15
digit_size = 28
grid_x_vae = np.linspace(-4+0.05, 4-0.05, n)
grid_y_vae = np.linspace(-4+0.05, 4-0.05, n)
VAE_figure = np.zeros((digit_size * n, digit_size * n))
for i, yi in enumerate(grid_x_vae):
    for j, xi in enumerate(grid_y_vae):
        z_sample = np.array([[xi, yi]])
        x_decoded = VAE_Decoder.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        VAE_figure[(n-i-1) * digit_size: (n - i) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = normalized(digit)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(VAE_figure, cmap='Greys')
plt.show()
```

```python
def inBetween(t, idx_1, idx_2):
    data_0 = x_train0[idx_1]
    data_1 = x_train0[idx_2]
    data_t = (1-t)*x_train0[idx_1] + t*x_train0[idx_2]

    mu_0 = VAE_Encoder.predict(x_train[idx_1:idx_1+1]).squeeze()
    mu_1 = VAE_Encoder.predict(x_train[idx_2:idx_2+1]).squeeze()
    mu_t = (1-t)*mu_0 + t*mu_1
    
    xhat_0 = VAE_Decoder.predict([[mu_0]]).reshape(28, 28)
    xhat_1 = VAE_Decoder.predict([[mu_1]]).reshape(28, 28)
    xhat_t = VAE_Decoder.predict([[mu_t]]).reshape(28, 28)

    plt.figure(figsize=(10, 5))
    ax1 = plt.subplot(3, 1, 3)
    ax1.scatter(mu_0[0], mu_0[1])
    ax1.scatter(mu_1[0], mu_1[1])
    ax1.scatter(mu_t[0], mu_t[1])

    ax2 = plt.subplot(3, 3, 1)
    ax2.imshow(data_0, cmap='Greys')

    ax3 = plt.subplot(3, 3, 2)
    ax3.imshow(data_t, cmap='Greys')

    ax4 = plt.subplot(3, 3, 3)
    ax4.imshow(data_1, cmap='Greys')
    
    ax5 = plt.subplot(3, 3, 4)
    ax5.imshow(xhat_0, cmap='Greys')

    ax6 = plt.subplot(3, 3, 5)
    ax6.imshow(xhat_t, cmap='Greys')

    ax7 = plt.subplot(3, 3, 6)
    ax7.imshow(xhat_1, cmap='Greys')
```

```python
interact(inBetween, t=FloatSlider(value=0.5, 
                                  min=0, 
                                  max=1.0,
                                  step=0.02,),
                    idx_1=IntSlider(value=30000, 
                                    min=0, 
                                    max=59999, 
                                    step=1,),
                    idx_2=IntSlider(value=30000, 
                                    min=0, 
                                    max=59999, 
                                    step=1,))
```
