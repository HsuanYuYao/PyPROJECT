```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = open('暑期生_數據組_7_22/data2.txt', "r")
x = df.read()
df.close()
```

```python
np.array(x)
x = x.split('   ')
y = []
for i in x:
    y.append(int(float(i)))
```

```python
# θ(0) = [Pa, Pb, ϵa, ϵb]
Pa = 0.3
Pb = 0.7
ϵa = 0.1
ϵb = 0.05
```

```python
# DP
z = [0]*1000
w = [0]*((1000-1)*2+1)
q = [0]*((1000-1)*2+1)

z[0] = 'A'
w[1] = (1-ϵa)*(Pa**(2-y[1])*(1-Pa)**(y[1]-1))
w[2] = ϵa*(Pb**(2-y[1])*(1-Pb)**(y[1]-1))
q[0] = 'A'
q[1] = 'A'
q[2] = 'A'

for i in 2*np.array(range(1, 1000-1))-1:
    a1 = w[i]*(1-ϵa)*(Pa**(2-y[int((i+1)/2)])*(1-Pa)**(y[int((i+1)/2)]-1))
    b1 = w[i+1]*ϵb*(Pa**(2-y[int((i+1)/2)])*(1-Pa)**(y[int((i+1)/2)]-1))
    if a1 > b1:
        w[i+2] = a1
        q[i+2] = 'A'
    else:
        w[i+2] = b1
        q[i+2] = 'B'
        
    a2 = w[i]*ϵa*(Pb**(2-y[int((i+1)/2)])*(1-Pb)**(y[int((i+1)/2)]-1))
    b2 = w[i+1]*(1-ϵb)*(Pb**(2-y[int((i+1)/2)])*(1-Pb)**(y[int((i+1)/2)]-1))
    if a2 > b2:
        w[i+3] = a2
        q[i+3] = 'A'
    else:
        w[i+3] = b2
        q[i+3] = 'B'
        
    if w[1997] > w[1998]:
        z[999] = 'A'
    else:
        z[999] = 'B'
        
for i in 1000-np.array(range(1, 1000)):
    if z[i] == 'A':
        z[i-1] = q[2*i-1]
    else:
        z[i-1] = q[2*i]
```

```python
n1 = 0
n2 = 0
n3 = 0
n4 = 0
m1 = 0
m2 = 0
m3 = 0
m4 = 0
for i in range(0, 200):
    if (y[i]==1) & (z[i]=='A'):
        n1 += 1
    elif (y[i]==2) & (z[i]=='A'):
        n2 += 1
    elif (y[i]==1) & (z[i]=='B'):
        n3 += 1
    else: # y[i]==2 & z[i]=='B'
        n4 += 1
    
    if i == 199:
        break
    elif (z[i]=='A') & (z[i+1]=='B'):
        m1 += 1
    elif (z[i]=='A') & (z[i+1]=='A'):
        m2 += 1
    elif (z[i]=='B') & (z[i+1]=='A'):
        m3 += 1
    else: # z[i]=='B' & z[i+1]=='B':
        m4 += 1

# Loss = P(X|Z, θ)P(Z|θ) = Pa**n1*(1-Pa)**n2*Pb**n3*(1-Pb)**n4 * ϵa**m1*(1-ϵa)**m2*ϵb**m3*(1-ϵb)**m4

# update θ
Pa = n1/(n1+n2)
Pb = n3/(n3+n4)
ϵa = m1/(m1+m2)
ϵb = m3/(m3+m4)
```

```python
print('z = ', z)
print(f'θ = [{Pa}, {Pb}, {ϵa}, {ϵb}]')
```
