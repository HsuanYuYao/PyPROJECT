```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = open('暑期生_數據組_7_22/data1.txt', "r")
x = df.read()
df.close()
```

```python
np.array(x)
x = x.split('   ')
x.remove('')
y = []
for i in x:
    y.append(float(i))
```

```python
# p = 1/2
μ1 = 50
σ1 = 3
μ2 = 60
σ2 = 2
```

```python
for j in range(0, 10):
    y1 = []
    y2 = []

    # Loss function
    # L = 1
    for i in y:
        # L = L*(p*(2*π*σ1**2)**(-1/2)*exp((i-μ1)**2/(2*σ1**2))+(1-p)*(2*π*σ2**2)**(-1/2)*exp((i-μ2)**2/(2*σ2**2)))
        # 分群
        f1 = (2*np.pi*σ1**2)**(-1/2)*np.exp(-(i-μ1)**2/(2*σ1**2))
        f2 = (2*np.pi*σ2**2)**(-1/2)*np.exp(-(i-μ2)**2/(2*σ2**2))

        if f1 > f2:
            y1.append(i)
        else:
            y2.append(i)

    # update θ
    μ1 = sum(y1)/len(y1)
    μ2 = sum(y2)/len(y2)

    y1 = np.array(y1)
    y2 = np.array(y2)
    σ1 = (sum((y1-μ1)**2)/len(y1))**(1/2)
    σ2 = (sum((y2-μ2)**2)/len(y2))**(1/2)

    print(f'[{μ1}, {μ2}, {σ1}, {σ2}]')
```

```python
for j in range(0, 50):
    
    y = np.array(y)
    z = []

    # Loss function
    # L = 1
    for i in y:
        # L = L*(p*(2*π*σ1**2)**(-1/2)*exp((i-μ1)**2/(2*σ1**2))+(1-p)*(2*π*σ2**2)**(-1/2)*exp((i-μ2)**2/(2*σ2**2)))
        f1 = (2*np.pi*σ1**2)**(-1/2)*np.exp(-(i-μ1)**2/(2*σ1**2))
        f2 = (2*np.pi*σ2**2)**(-1/2)*np.exp(-(i-μ2)**2/(2*σ2**2))

        z.append(f1/(f1+f2))

    # update θ
    
    z = np.array(z)
    μ1 = sum(y*z)/sum(z)
    μ2 = sum(y*(1-z))/sum(1-z)

    σ1 = (sum(z*(y-μ1)**2)/sum(z))**(1/2)
    σ2 = (sum((1-z)*(y-μ2)**2)/sum(1-z))**(1/2)

    print(f'[{μ1}, {μ2}, {σ1}, {σ2}]')
```
